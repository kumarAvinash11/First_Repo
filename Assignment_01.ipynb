```python
#Import the required library
import seaborn as sns
import pandas as pd
```

#### 2: Write a user-defined function for origin  


```python
#use apply function
def origin(num):
    if num ==1:
        return 'IND'
    if num==2:
        return 'USA'
    if num==3:
        return 'JAPAN'


auto_data['origin'] = auto_data['origin'].apply(origin)
```

```python
#view first 30 data points 
auto_data.head(30)
```

#### 3: Draw the pair plot using sns for mpg, weight,orign and with hue origin, set the size to 4


```python
df_new_auto = auto_data[['mpg', 'weight', 'origin']]
```

```python
sns.pairplot(df_new_auto, hue ='origin', size = 4)
```

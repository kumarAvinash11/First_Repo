#### 1: List the causes of death and percentile in Ohio state in the year 2012

```python
#import required libraries
#import pandas as pd
import matplotlib.pyplot as plt
```

```python
# View the plot 
%matplotlib inline
```

```python
#cause of deaths
cause_of_deaths = ['Chronic Disease',' Unintetional Injuries', 'Alzheimers',' Infuenza and Pneumonia', 'Sepsis','Others']
#percentile
percentile_of_death = [62, 5, 4, 2, 1 , 26 ]
```

#### 2: Draw a pie chart using the two data points

```python
#set the pie chart plot properties
```

```python
#set figure size
plt.figure(figsize = (10,10))
```

```python
#explode the largest pie in the data set
explode = [0.05,0,0,0,0,0]
```

```python
#set pie chart properties
plt.pie(percentile_of_death, labels= cause_of_deaths, explode = explode, autopct ='%1.1f%%', startangle = 90)
```

```python
#set axis equal to draw pie as circle
plt.axis('equal')
```

```python
#set title of the pie chart
plt.title('Ohio_state_death')
```

```python
#show the plot (pie chart)
plt.show()
```
